```python

import pandas as pd

# Load with header row
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
dataset = pd.read_csv("pima-indians-diabetes.csv", header=0, names=col_names)

# Remove any rows where column names are present as data
dataset = dataset[dataset['pregnant'] != 'pregnant']

# Convert all columns to numeric (if needed)
for col in col_names:
    dataset[col] = pd.to_numeric(dataset[col], errors='coerce')

# Now split features and target
feature_cols = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']
X = dataset[feature_cols]
y = dataset['label']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn import tree

# 1. Load dataset
pima = pd.read_csv("pima-indians-diabetes.csv")

# Check columns
print("Columns in dataset:", pima.columns.tolist())

# 2. Features and target (use exact column names from CSV)
feature_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness',
                'Insulin','BMI','DiabetesPedigreeFunction','Age']

X = pima[feature_cols]     # Feature matrix
y = pima['Outcome']        # Target column

# 3. Split dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

# 4. Create Decision Tree classifier
clf = DecisionTreeClassifier()

# 5. Train the classifier
clf = clf.fit(X_train, y_train)

# 6. Predictions
y_pred = clf.predict(X_test)

# 7. Accuracy
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))

# 8. Plot decision tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf,
               feature_names=feature_cols,
               class_names=['No Diabetes','Diabetes'],
               filled=True,
               rounded=True)
plt.show()
```

```python

print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("f1_Score:",metrics.f1_score(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn import tree

# 1. Load dataset
pima = pd.read_csv("C:/Users/ABHISHEK/Downloads/study material/5th sem/Machine Learning/diabetes.csv")

# 2. Features and target
feature_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness',
                'Insulin','BMI','DiabetesPedigreeFunction','Age']

X = pima[feature_cols]
y = pima['Outcome']

# 3. Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

# -----------------------
# Full-depth Decision Tree
# -----------------------
clf_full = DecisionTreeClassifier(random_state=1)
clf_full = clf_full.fit(X_train, y_train)

plt.figure(figsize=(15,10))
tree.plot_tree(clf_full,
               feature_names=feature_cols,
               class_names=['No Diabetes','Diabetes'],
               filled=True, rounded=True)
plt.title("Full Decision Tree")
plt.show()

# -----------------------
# Limited-depth Decision Tree (4 levels)
# -----------------------
clf_4 = DecisionTreeClassifier(max_depth=4, random_state=1)
clf_4 = clf_4.fit(X_train, y_train)

plt.figure(figsize=(15,10))
tree.plot_tree(clf_4,
               feature_names=feature_cols,
               class_names=['No Diabetes','Diabetes'],
               filled=True, rounded=True)
plt.title("Decision Tree (Max Depth = 4)")
plt.show()

# Accuracy check
y_pred_4 = clf_4.predict(X_test)
print("Accuracy with max_depth=4:", metrics.accuracy_score(y_test, y_pred_4))
```
