```python
import matplotlib.pyplot as plt
import seaborn as sns

# Single bar plot for a categorical column (e.g., Fuel_Type)
plt.figure(figsize=(8, 5))
sns.countplot(x='Fuel_Type', data=data, color='skyblue', order=data['Fuel_Type'].value_counts().index)
plt.title('Count of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("used_cars_data.csv")

df['Mileage'] = df['Mileage'].str.extract(r'(\d+\.\d+|\d+)').astype(float)


df = df[['Price', 'Mileage']].dropna()

plt.figure(figsize=(8, 5))
plt.scatter(df['Mileage'], df['Price'], alpha=0.10, color='skyblue')
plt.xlabel('Mileage (kmpl or km/kg)')
plt.ylabel('Price (in lakhs)')
plt.title('Scatter Plot: Price vs Mileage')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("used_cars_data.csv")


current_year = 2025
df['Age'] = current_year - df['Year']


df = df[['Price', 'Age']].dropna()


plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Price'], alpha=0.25, color='red')
plt.xlabel('Age of Car (in years)')
plt.ylabel('Price (in lakhs)')
plt.title('Scatter Plot: Price vs Age')
plt.grid(True)
plt.tight_layout()
plt.show()
```
